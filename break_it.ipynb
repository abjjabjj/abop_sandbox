```python
import concrete_async
import concrete_db
from concrete_core.orchestrator import SoftwareOrchestrator
from concrete_core.clients import CLIClient
```

```python
async def main(prompt: str, store_messages: bool=False, run_async: bool=False):
    so = SoftwareOrchestrator(store_messages)
    CLIClient.emit("")
    async for operator, response in so.process_new_project(prompt, run_async=run_async):
        CLIClient.emit(operator)
        CLIClient.emit(response)
```

```python
await main("Write a coin flipper", store_messages=True, run_async=True)
```
